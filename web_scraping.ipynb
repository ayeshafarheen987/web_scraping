# Scraping all the clinics from Trustpilot

```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.trustpilot.com/categories/clinics')
```

```python
response
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())[:100]
```

```python
clinics = soup.select('div.styles_logoBox__SF4DX~div.styles_businessUnitMain__PuwB7')
```

```python
len(clinics)
```

```python
clinics
```

```python
print(clinics[0].prettify())
```

```python
url = 'https://www.trustpilot.com/categories/clinics'
```

```python
total_num = 17
clinics = []
for page_num in range(1,total_num+1):
    tmp_url = url+'?page'+str(page_num)
    response = requests.get(tmp_url)
    soup = BeautifulSoup(response.content,'html.parser')
    tmp_clinics = soup.select('div.styles_logoBox__SF4DX~div.styles_businessUnitMain__PuwB7')
    print(f'page #{page_num}:\t{len(tmp_clinics)}results found...')
    clinics.extend(tmp_clinics)
```

```python
len(clinics)
```

```python
len(tmp_clinics)
```

```python
print(clinics[-1].prettify())
```

```python
print(clinics[0].prettify())
```

```python
name = clinics[0].select('p.styles_displayName__GOhL2')[0].text
score = clinics[0].select('span.styles_trustScore__8emxJ')[0].text
location = clinics[0].select('span.styles_location__ILZb0')[0].text
reviews = clinics[0].select('p.styles_ratingText__yQ5S7')[0].text.split('|')[-1]
name, score, location, reviews
```

```python
data = {'name':[],
       'score':[],
       'location':[],
       'review':[]
       }
for i in clinics:
    data['name'].append(i.select('p.styles_displayName__GOhL2')[0].text)
    
    try:
        data['score'].append(i.select('span.styles_trustScore__8emxJ')[0].text)
    except:
        data['score'].append(None)
        
    try:
        data['location'].append(i.select('span.styles_location__ILZb0')[0].text)
    except:
        data['location'].append(None)
        
    try:
        data['review'].append(i.select('p.styles_ratingText__yQ5S7')[0].text)
    except:
        data['review'].append(None)
        
```

```python
df = pd.DataFrame(data)
```

# Scraping all the reviews from a clinic in Trustpilot

```python
url = 'https://www.trustpilot.com/review/relaxedclarity.com'
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
reviews = soup.select('article')
```

```python
len(reviews)
```

```python
print(reviews[-1].prettify())
```

```python
title = reviews[0].select_one('h2.typography_heading-s__f7029').text
time = reviews[0].select_one('time').text
description = reviews[0].select_one('p.typography_body-l__KUYFJ').text
rating = reviews[0].select_one('img')['alt']
```

```python
title
```

```python
time
```

```python
description
```

```python
rating
```

```python
reviews = []
num_pages = 173
for i in range(1,num_pages+1):
    tmp_url = url+'?page'+str(num_pages)
    response = requests.get(tmp_url)
    soup = BeautifulSoup(response.content,'html.parser')
    tmp_reviews = soup.select('article')
    print(f'page #{num_pages}:\t{len(tmp_reviews)}result found..')
    reviews.extend(tmp_reviews)
```

```python
data = {
    'title' : [],
    'description':[],
    'time':[],
    'rating':[]
}
for i in reviews:
    try:
        data['title'].append(i.select_one('h2.typography_heading-s__f7029').text)
    except:
        data['title'].append(None)
        
    try:
        data['description'].append(i.select_one('p.typography_body-l__KUYFJ').text)
    except:
        data['description'].append(None)
        
    try:
        data['time'].append(i.select_one('time').text)
    except:
        data['time'].append(None)
        
    try:
        data['rating'].append(i.select_one('img')['alt'])
    except:
        data['rating'].append(None)
```

```python
df = pd.DataFrame(data)
```
